# 1. Write a Python program to find a target values in a list using linear search with following steps:

# 2. Write a Python program to implement binary search to find the target values from the list: 
a. Create a separate function to do binary search. b. Get the number of inputs from the user. c. Store the inputs individually in a list. d. In binary search function at first sort the list in order to start the search from middle of the list. e. Compare the middle element to right and left elements to search target element. f. If greater, move to right of list or else move to another side of the list. g. Print the result along with the position of the element. 


```python
#--------------- Define a function for binary search ---------------#

def binary_search(arr, target):
    
   #--------------- Sort the list in ascending order ---------------#
    
    arr.sort()
    
    left, right = 0, len(arr) - 1

    while left <= right:
        
        mid = left + (right - left) // 2

        if arr[mid] == target:
            
            return mid  #--------------- Target found, return its index ---------------#

        if arr[mid] < target:
            
            left = mid + 1
            
        else:
            
            right = mid - 1

    return -1  #--------------- Target not found in the list ---------------#


#--------------- Input the number of elements from the user with error handling ---------------#

while True:
    
    try:
        
        num_elements = int(input("\nEnter the number of elements in the list: "))
        
        break 
        
    except ValueError:
        
        print("\nInvalid input. Please enter a valid integer.")

        
#--------------- Create an empty list to store the elements ---------------#

input_elements = []


#--------------- Input the elements individually and populate the list with error handling ---------------#

for i in range(num_elements):
    
    while True:
        
        try:
            
            element = int(input(f"\nEnter element {i+1}: "))
            
            input_elements.append(element)
            
            break  
            
        except ValueError:
            
            print("\nInvalid input. Please enter a valid integer.")

            
#--------------- Input the target value to search with error handling ---------------#

while True:
    
    try:
        
        target_value = int(input("\nEnter the value to search from the list: "))
        
        break  
        
    except ValueError:
        
        print("\nInvalid input. Please enter a valid integer.")
        

#--------------- Call the binary_search function to find the target value ---------------#

result = binary_search(input_elements, target_value)


#--------------- Display the result ---------------#

if result != -1:
    
    print(f"\n{target_value} found at position {result}")
    
else:
    
    print(f"\n{target_value} is not found in the list.")
```

# 3. Write a Python program for sorting a list of elements using selection sort algorithm: 
a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list. b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted list. c. In the second iteration, find the smallest element in the unsorted list and place it in the correct position by comparing with the element in the sorted list. d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position by comparing with the elements in the sorted list. e. This process continues till the unsorted list becomes empty. f. Display the sorted list. 


```python
def selection_sort(unsorted_list):
    
   #--------------- Initialize an empty sorted list ---------------#
    
    sorted_list = []  
    
    while unsorted_list:
        
        #--------------- Find the minimum value in the unsorted list ---------------#
        
        min_value = min(unsorted_list)   
        
       #--------------- Remove the minimum value from the unsorted list ---------------#
        
        unsorted_list.remove(min_value)  

       #--------------- Place the minimum value in the correct position in the sorted list ---------------#
        
        for i in range(len(sorted_list)):
            
            if min_value < sorted_list[i]:
                
                sorted_list.insert(i, min_value)
                
                break
        else:
            
          #--------------- If it's the largest, append it to the end ---------------#
            
            sorted_list.append(min_value)  

    return sorted_list



#--------------- Input a list of elements from the user ---------------#

input_list = input("\nEnter a list of elements: ").split()

input_list = [int(item) for item in input_list]



#--------------- Sort the list using selection sort ---------------#

sorted_list = selection_sort(input_list)



#--------------- Display the sorted list ---------------#

print("\nSorted List elements:", sorted_list)
```

# 4. Write a Python program for sorting a list of elements using insertion sort algorithm: 
a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list. b. In the first iteration, take the first element in the unsorted list and insert it in Sorted list. c. In the second iteration, take the second element in the given list and compare with the element in the sorted sub list and place it in the correct position. d. In the third iteration, take the third element in the given list and compare with the elements in the sorted sub list and place the elements in the correct position. e. This process continues until the last element is inserted in the sorted sub list. f. Display the sorted elements. 


```python
# Step a: Assume two lists:

# Sorted list - Initially empty

sorted_list = []

# Unsorted List - Given input list

unsorted_list = [10, 4, 8, 6, 15]

# Display the unsorted list

print("\nUnsorted List:", unsorted_list)

# Step b: In the first iteration, take the first element in the unsorted list and insert it in the sorted list.

sorted_list.append(unsorted_list[0])

unsorted_list = unsorted_list[1:]

# Step c, d, e: Continue inserting elements from the unsorted list into the sorted list while maintaining sorting order.

for i in range(len(unsorted_list)):
    
    current_element = unsorted_list[i]
    
    j = len(sorted_list) - 1
    
    while j >= 0 and current_element < sorted_list[j]:
        
        j -= 1
        
    sorted_list.insert(j + 1, current_element)


# Step f: Display the sorted elements.

print("\nSorted List:", sorted_list)
```

# 5. Write a Python program that performs merge sort on a list of numbers: 
a. Divide: If the given array has zero or one element, return. 1. Otherwise ii. Divide the input list in to two halves each containing half of the elements. i.e. left half and right half. b. Conquer: Recursively sort the two lists (left half and right half). a. Call the merge sort on left half. b. Call the merge sort on right half. c. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence. 


```python
def merge_sort(arr):
    
    if len(arr) <= 1:
        
        return arr  

    
    #--------------- Calculate the middle index to split the list into two halves ---------------#
    
    mid = len(arr) // 2

    
    #--------------- Divide the input list into left and right halves ---------------#
    
    left_half = arr[:mid]
    
    right_half = arr[mid:]

    
    #--------------- Recursively sort the left and right halves ---------------#
    
    left_half = merge_sort(left_half)
    
    right_half = merge_sort(right_half)

    
    
    #--------------- Combine the two sorted halves into a single sorted list ---------------#
    
    return merge(left_half, right_half)



def merge(left, right):
    
    merged = []
    
    left_idx, right_idx = 0, 0

    while left_idx < len(left) and right_idx < len(right):
        
        if left[left_idx] < right[right_idx]:
            
            merged.append(left[left_idx])
            
            left_idx += 1
            
        else:
            
            merged.append(right[right_idx])
            
            right_idx += 1

            
            
    #--------------- Append any remaining elements from the left and right lists ---------------#
    
    merged.extend(left[left_idx:])
    
    merged.extend(right[right_idx:])

    return merged



#--------------- Input a list of numbers from the user ---------------#

input_list = input("Enter a list of numbers separated by spaces: ").split()

input_list = [int(item) for item in input_list] 



#--------------- Call the merge_sort function to sort and divide the list ---------------#

sorted_list = merge_sort(input_list)


#--------------- Display the sorted list ---------------#

print("\nSorted List:", sorted_list)
```

# 6. Write a Python script to perform the following operations on a singly linked list.
a. Create a list b. Find the smallest element from the list c. Insert an element if it is not a duplicate element d. Display the elements in reverse order 


```python
class Node:
    
    def __init__(self, value):
        
        self.value = value
        
        self.next_node = None
        
        

class SimpleLinkedList:
    
    def __init__(self):
        
        self.head = None

    def append(self, value):
        
        new_node = Node(value)
        
        if self.head is None:
            
            self.head = new_node
            
        else:
            
            current = self.head
            
            while current.next_node:
                
                current = current.next_node
                
            current.next_node = new_node
            
            

    def display(self):
        
        if self.head is None:
            
            print("\nEmpty List.")
            
        else:
            
            current = self.head
            
            while current:
                
                print(current.value, end=" -> ")
                
                current = current.next_node
                
            

            
    def find_smallest(self):
        
        if self.head is None:
            
            return None
        
        current = self.head
        
        smallest = current.value

        while current:
            
            if current.value < smallest:
                
                smallest = current.value
                
            current = current.next_node

        return smallest

    
    
    def insert_if_not_duplicate(self, value):
        
        if self.head is None:
            
            self.head = Node(value)
            
            return

        current = self.head
        
        while current:
            
            if current.value == value:
                
                return
            
            if current.next_node is None:
                
                current.next_node = Node(value)
                
                return
            
            current = current.next_node
            
            

    def display_reverse(self):
        
        if self.head is None:
            
            print("\nEmpty List.")
            
        else:
            
            values = []
            
            current = self.head
            
            while current:
                
                values.append(current.value)
                
                current = current.next_node
                
            values.reverse()
            
            for value in values:
                
                print(value, end=" -> ")
                
            
            

#--------------- Create a linked list and add elements ---------------#

my_linked_list = SimpleLinkedList()

my_linked_list.append(5)

my_linked_list.append(2)

my_linked_list.append(8)



#--------------- Insert an element only if it's not a duplicate ---------------#

my_linked_list.insert_if_not_duplicate(2)  # This won't insert because 2 is already in the list

my_linked_list.insert_if_not_duplicate(3)  # This will insert 3 because it's not a duplicate



#--------------- Display the modified linked list ---------------#

print("\nLinked List:")

my_linked_list.display()



#--------------- Display the elements in reverse order ---------------#
print("\nLinked List in Reverse:")

my_linked_list.display_reverse()
```

# 7. Write a python program to implement the various operations for Stack ADT 
i.) Push ii.) Pop iii.) Display. 


```python
class Stack:
    
    def __init__(self):
        
        self.items = []

        
        
    def push(self, item):
        
        self.items.append(item)

        
        
    def pop(self):
        
        if not self.is_empty():
            
            return self.items.pop()

        
        
    def is_empty(self):
        
        return len(self.items) == 0

    
    
    def display(self):
        
        print("Stack contents:")
        
        for item in reversed(self.items):
            
            print(item)

            
            
#--------------- create a new stack object ---------------#

stack = Stack()



#--------------- push elements onto the stack ---------------#

stack.push(10)
stack.push(20)
stack.push(30)



#--------------- display the contents of the stack ---------------#

stack.display()



#--------------- pop an element from the stack ---------------#

popped_item = stack.pop()

if popped_item is not None:
    
    print("Popped item:", popped_item)

    
    
#--------------- Display Stack contents ---------------#

stack.display()
```

# 8. Write a python script to implement the various operations for Queue ADT 
i.) Insert ii.) Delete iii.) Display. 


```python
class Queue:
    
    def __init__(self):
        
        self.items = []

        
        
    def is_empty(self):
        
        return len(self.items) == 0

    
    
    def insert(self, item):
        
        self.items.append(item)

        
        
    def delete(self):
        
        if not self.is_empty():
            
            return self.items.pop(0)
        
        else:
            
            print("\nQueue is empty. Cannot delete.")

            
            
    def display(self):
        
        if not self.is_empty():
            
            print("\nQueue elements (front to rear):", self.items)
            
        else:
            
            print("\nQueue is empty.")


            
            
#--------------- Create an empty queue ---------------#

my_queue = Queue()



#--------------- Insert elements into the queue ---------------#

my_queue.insert(200)

my_queue.insert(400)

my_queue.insert(600)



#--------------- Display the queue ---------------#

print("\nQueue after inserts:")

my_queue.display()



#--------------- Delete an element from the queue ---------------#

deleted_item = my_queue.delete()

if deleted_item is not None:
    
    print(f"\nDeleted element: '{deleted_item}'")

    
    
#--------------- Display the queue again ---------------#

print("\nQueue after delete operation:")

my_queue.display()
```

# 9. Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack 

a. A/B^C+D*E-F*G 

b. (B^2-4*A*C)^(1/2) (100) 

```python
#--------------- Function to check for an operator ---------------#


def is_operator(char):
    
    return char in "+-*/^"



#--------------- Function to get the precedence of an operator ---------------#


def get_precedence(char):
    
    if char == '^':
        
        return 3
    
    elif char in "*/":
        
        return 2
    
    elif char in "+-":
        
        return 1
    
    else:
        
        return 0


    
##--------------- Function to convert an infix expression to postfix ---------------#


def infix_to_postfix(infix):
    
    stack = []
    
    postfix = []
    
    for char in infix:
        
        if char.isalnum():
            
            postfix.append(char)
            
        elif char == '(':
            
            stack.append(char)
            
        elif char == ')':
            
            while stack and stack[-1] != '(':
                
                postfix.append(stack.pop())
                
            if stack and stack[-1] == '(':
                
                stack.pop()
                
        elif is_operator(char):
            
            while stack and get_precedence(char) <= get_precedence(stack[-1]):
                
                postfix.append(stack.pop())
                
            stack.append(char)
            
    while stack:
        
        postfix.append(stack.pop())
        
    return ''.join(postfix)



#--------------- User Inputs ---------------#

infix1 = "A/B^C+D*E-F*G"

infix2 = "(B^2-4*A*C)^(1/2) (100)"

postfix1 = infix_to_postfix(infix1)

postfix2 = infix_to_postfix(infix2)

print("\nInfix expression 1:", infix1)

print("\nPostfix expression 1:", postfix1)

print("\nInfix expression 2:", infix2)

print("\nPostfix expression 2:", postfix2)
```
